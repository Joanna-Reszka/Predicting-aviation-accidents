### EDA
-  visualisation of the effect of each column on target column

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('df_US_ml.csv')
```

```python
df.info()
```

```python
df.columns
```

```python
palette1 = {
    'Minor_Uninjured': 'tab:blue',
    'Fatal_Injured': 'black',
}
```

### Aircraft Category

```python
plt.figure(figsize = (6,3))
sns.countplot(data=df, x= 'Aircraft.Category', hue = 'Target_binary', palette=palette1) 
```

### Amateur Build

```python
plt.figure(figsize = (6,3))
sns.countplot(data=df, x= 'Amateur.Built', hue = 'Target_binary', palette=palette1)  
```
